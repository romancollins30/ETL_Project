```python
import pandas as pd
pd.set_option('max_rows',1000) #To display 1000 rows
```

```python
url = 'https://www.spotrac.com/mlb/rankings/'
```

```python
table = pd.read_html(url)
table
```

```python
type(table)
```

```python
df = table[0]
df.columns = ['Id', 'Player', 'Position', 'Salary']
df.head()
```

```python
df = df.drop(['Player'], axis=1)
```

```python
# new data frame with split value columns 
new = df["Position"].str.split(" ", n = 1, expand = True) 
  
# making seperate first name column from new data frame 
df["First Name"]= new[0] 
  
# making seperate last name column from new data frame 
df["Last Name"]= new[1] 
  
# df display 
df
```

```python
# new data frame with split value columns 
new = df["Position"].str.split(" ", n = 1, expand = True) 
  
# making seperate first name column from new data frame 
df["Drop"]= new[0] 
  
# making seperate last name column from new data frame 
df["Keep"]= new[1] 
  
# df display 
df
```

```python
# new data frame with split value columns 
new = df["Keep"].str.split(" ", n = 1, expand = True) 
  
# making seperate first name column from new data frame 
df["Drop2"]= new[0] 
  
# making seperate last name column from new data frame 
df["Keep2"]= new[1] 
  
# df display 
df
```

```python
df = df.drop(['Position', 'Drop'], axis=1)
df
```

```python
df = df.drop(['Last Name', 'Keep'], axis=1)
df
```

```python
df['Position'] = df['Keep2']
df
```

```python
df = df.drop(['Keep2'], axis=1)
df
```

```python
df = df.drop(['Player'], axis=1)
df
```

```python
df["name"] = df["First Name"].map(str) + " " + df["Drop2"]
df
```

```python
df = df.drop(['First Name', 'Drop2'], axis=1)
df
```

```python
df = df[['name','Position','Salary']]
df
```

```python
df.to_csv('2018_MLB_Salaries.csv')
```
