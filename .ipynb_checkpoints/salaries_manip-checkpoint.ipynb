```python
import pandas as pd
```

```python
people = 'people.csv'
people_df = pd.read_csv(people)
people_df
```

```python
people_df['playerID'].dtype
```

```python
salaries = 'salaries.csv'
sal_df = pd.read_csv(salaries)
sal_df
```

```python
join_df = pd.merge(sal_df, people_df,how='left', on='playerID')
join_df = join_df[['nameFirst', 'nameLast', 'teamID', 'yearID', 'salary', ]]
join_df = join_df.loc[join_df['yearID'] >= 2007, :]
join_df = pd.DataFrame(join_df)
```

```python
join_df
```

```python
join_df = join_df.reset_index(drop=True)
join_df.to_csv('old_salaries.c')
```
